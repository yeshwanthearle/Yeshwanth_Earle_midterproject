```python
import pandas as pd
import numpy as np
import os
import sys
from itertools import combinations, permutations
```

```python
#reading data based on the user input
print("1.Amazon Transactions\t2.Best Buy Transactions\t3.Nike Transactions\t4.KMart Transactions\t5.Custom Transactions")
user_choice = input("Please Enter your choice of data:")
try:
    user_choice = int(user_choice)
    if not 0<=user_choice<=6:
        sys.exit("Invalid Input!")
    data_file = ""
    if user_choice == 1:
        data_file = "amazon.csv"
    elif user_choice == 2:
        data_file = "bestbuy.csv"
    elif user_choice == 3:
        data_file = "nike.csv"
    elif user_choice == 4:
        data_file = "kmart.csv"
    else:
        data_file = "1.csv"
    data =  pd.read_csv(data_file)
except ValueError:
    sys.exit("Invalid Input!")
print("entered data", data_file)
```

```python
col = list(data.columns)[1]
item_sets = list(data[col].str.split(',').values)
items = dict()
for item_set in item_sets:
    for item in item_set:
        if item:
            item = item.strip()
            try:
                items[item] = items[item] + 1
            except:
                items[item] = 1
                
print(items)
```

```python
for item in items.keys():
    data[item] = 0
data.head()
```

```python
for row in range(data.shape[0]):
    prods = data[col][row]
    prods = prods.split(',')
    for prod in prods:
        if prod:
            data[prod.strip()][row] = 1
data.head()
```

```python
min_support = input("Enter Minimum Support(as a value between 0 and 1):")
try:
    min_support = float(min_support)
    if not 0<=min_support<=1:
        print("Invalid Input!")
        sys.exit()
except ValueError:
    print("Invalid Input!")
    sys.exit()
print("Minimum support:",float(min_support))
```

```python
min_confidence = input("Enter Minimum Confidence(as a value between 1 and 100):")
try:
    min_confidence = float(min_confidence)
    if not 1<=min_confidence<=100:
        print("Invalid Input!")
        sys.exit()
except ValueError:
    print("Invalid Input!")
    sys.exit()
print("Minimum Confidence %:",int(min_confidence))
```

```python
items_with_min_support = {}
for item in items:
    total = np.sum(data[item])/data.shape[0]
    if total>=min_support:
        items_with_min_support[item] = total
print(items_with_min_support)
```

```python
frequent_item_sets = {}
min_support_item_count = len(items_with_min_support.keys())
for i in range(2, min_support_item_count+1):
    trans_sets = list(combinations(list(items_with_min_support.keys()), i))
    for tran in trans_sets:
        print(list(tran))
        support = np.sum(np.sum(data[list(tran)], axis=1)==i)
        if  support > int(min_support):
            frequent_item_sets[tran] = support
print(frequent_item_sets)
```

```python
results = []
for item_set in frequent_item_sets.keys():
    trans = list(permutations(item_set))

    for tran in trans:
        tran = list(tran)
        support_item_set = np.sum(np.sum(data[tran], axis=1)==len(tran))
        support_item_sub_set = np.sum(np.sum(data[tran[:-1]], axis=1)==len(tran)-1)
        confidence = support_item_set/support_item_sub_set*100

        if  confidence >= min_confidence:
            results.append((tran, str(round(confidence,1))+"%"))
```

```python
if results:
    for result in results:
        items = ', '.join(result[0][:-1])
        print("Item set:", result[0])
        print('-'*100)
        print("With Confidence of", result[1], "we can say that if a customer buys ", items ,"then he/she will also buy ", result[0][-1])
        print("*"*100)
else:
    print("No Item sets found with given support and confidence conditions")
```
